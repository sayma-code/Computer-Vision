```python
import cv2
from matplotlib import pyplot as plt
img_bgr = cv2.imread('example.png', cv2.IMREAD_COLOR)
# cv2.imread loads the file example.png from disk as a NumPy array.
# cv2.IMREAD_COLOR tells OpenCV to read a color image with 3 channels in BGR order (Blue, Green, Red).
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
img_rgb[10,16] = [255,255,255]
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

32 * 32 * 3 = 3072 , 3 kilobytes

```python
img_rgb_dark = img_rgb // 2
plt.imshow(img_rgb_dark)
plt.axis("off")
plt.show()
```

```python
import numpy as np
img_rot = np.zeros(shape=(32, 32, 3), dtype=np.uint8)

for j in range(0,32):
  for i in range(0,32):
    for c in range(0,3):
      img_rot[j,i,c] = img_bgr[i,j,c]
plt.imshow(img_rot)
plt.axis("off")
plt.show()
```
